# Data Preprocessing

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('data.csv') # data frame
# Country Age Salary Purchased

# iloc = locate indexes
X = dataset.iloc[:, :-1].values # matrix of features, independent variables
y = dataset.iloc[:, -1].values # dependent variable vector
```

```python
X
# Country Age Salary
```

```python
y
# Purchased
```

## Missing Data

If the missing data is 1% or less, remove those rows.

Strategy: Replace missing data with the average of the column.

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean') # or 'median', 'most_frequent'
imputer.fit(X[:, 1:3]) # exclude Country column, select numerical columns
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```python
X
```

## Encode Categorical Data

### Encode Independent Variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
# 'transformers' is list of (name, transformer, indexes of columns) tuples. [()]
# 'remainder' is 'passthrough', means we want to keep the columns that we did not apply transformation

# Country column
X = np.array(ct.fit_transform(X))
```

```python
X
```

### Encode Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

## Split Data into Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# Do not apply feature scaling to dummy variables
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```
